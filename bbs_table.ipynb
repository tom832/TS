```python
import pandas as pd
from rdkit import Chem
from rdkit.Chem import Draw
from openpyxl import Workbook
from openpyxl.drawing.image import Image
from io import BytesIO

# 读取CSV数据
data = pd.read_csv('/home/jnliu/ugi/ugi_synthesis/20240501.csv')

# 创建一个新的Excel工作簿
wb = Workbook()
ws = wb.active

# 设置标题行
ws.append(['Molecule Image', 'ID', 'CAS', 'Cluster ID'])

# 图像尺寸
img_width, img_height = 300, 300

# 为每个分子生成图像并写入Excel
for index, row in data.iterrows():
    # 使用RDKit从SMILES字符串生成分子对象
    mol = Chem.MolFromSmiles(row['uc_smiles'])
    
    # 生成分子的图像
    img = Draw.MolToImage(mol, size=(img_width, img_height))
    
    # 将PIL图像转换为Excel可以插入的格式
    img_byte_arr = BytesIO()
    img.save(img_byte_arr, format='PNG')
    img_byte_arr = BytesIO(img_byte_arr.getvalue())
    
    # 在Excel中插入图像
    img_excel = Image(img_byte_arr)
    
    # 将图像和其他数据添加到工作簿
    ws.append([None, row['mol_id'], row['cas'], row['cluster_id']])
    img_cell = 'A' + str(index + 2)  # 计算图像应该插入的单元格位置
    ws.add_image(img_excel, img_cell)
    
    # 调整行高以适应图像
    ws.row_dimensions[index + 2].height = img_height * 0.75  # Excel行高单位与像素不同，可能需要调整

# 保存Excel文件
wb.save('output_random.xlsx')
```
