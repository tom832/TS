```python

# 数据
# labels = ['positive', 'negative']
# sizes = [2, 22]

# # 创建饼图
# fig, ax = plt.subplots()
# plt.title("random select")
# wedge_properties = {'alpha': 0.6}
# ax.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, wedgeprops=wedge_properties)

# # 等比例显示，确保饼图为圆形
# ax.axis('equal')

# # 显示饼图
# plt.show()

import matplotlib.pyplot as plt

labels = ['positive', 'negative']
sizes = [2, 22]
# colors = ['lightblue', 'lightgreen']  # 设置颜色
ring_width = 0.3  # 设置环的宽度
pct_distance = 0.4  # Increase the distance for percentage labels

# 创建环形图
fig, ax = plt.subplots()
plt.title("Random Selection", fontsize=18, y=1.08)  # Move the title up a bit
wedge_properties = {'alpha': 0.6, 'width': ring_width}
text_properties = {'fontsize': 18}  # 设置字体大小

# 自定义显示格式，同时显示百分比和实际数值
def autopct_format(values):
    def inner_autopct(pct):
        total = sum(values)
        val = int(round(pct*total/100.0))
        return f'{pct:.1f}% ({val})'
    return inner_autopct

ax.pie(sizes, labels=labels, autopct=autopct_format(sizes), pctdistance=pct_distance, startangle=90, wedgeprops=wedge_properties, textprops=text_properties)

# 等比例显示，确保饼图为圆形
ax.axis('equal')

# 显示环形图
plt.show()
```
